```python
df=pd.read_hdf('Week_Number.h5','a')
df['Avg_Rank']=df[['Rank_First_Week','Rank_Second_Week','Rank_Third_Week','Rank_Forth_Week']].mean(axis=1)
df['Avg_First_Week']=df['First_Week']/4
df['Avg_Second_Week']=df['Second_Week']/7
df['Avg_Third_Week']=df['Third_Week']/7
df['Avg_Forth_Week']=df['Forth_Week']/7
df['Avg_Final_Week']=df['Final_Week']/5
df['Total_Number']=df[['First_Week','Second_Week','Third_Week','Forth_Week']].sum(axis=1)
df['Rank_number']=df['Total_Number'].rank(ascending=False,method='dense')
df['Rank_number']=df['Rank_number'].astype(str)
df['Rank_number']=df['Rank_number'].apply(lambda n:n[:n.index('.')])
df['Rank_number']=df['Rank_number'].astype('int64')
df['Avg_Number']=df.Total_Number/4
df['Max_Rank']=df[['Rank_First_Week','Rank_Second_Week','Rank_Third_Week','Rank_Forth_Week']].max(axis=1)
df['Min_Rank']=df[['Rank_First_Week','Rank_Second_Week','Rank_Third_Week','Rank_Forth_Week']].min(axis=1)
df['Avg_mid_Rank']=df[['Max_Rank','Min_Rank']].mean(axis=1)
df['sd']=df[[ 'Rank_First_Week', 'Rank_Second_Week', 'Rank_Third_Week','Rank_Forth_Week']].std(axis=1,ddof=1)
df['median']=df[[ 'Rank_First_Week', 'Rank_Second_Week', 'Rank_Third_Week','Rank_Forth_Week']].median(axis=1)
df['mad']=df[[ 'Rank_First_Week', 'Rank_Second_Week', 'Rank_Third_Week','Rank_Forth_Week']].mad(axis=1)
df['variance']=df[[ 'Rank_First_Week', 'Rank_Second_Week', 'Rank_Third_Week','Rank_Forth_Week']].var(axis=1,ddof=1)
df['First_Rank_Gap']=df['Rank_First_Week'].apply(lambda n:1 if n<=100 else 2 if n<=1000 else 3)
df['Second_Rank_Gap']=df['Rank_Second_Week'].apply(lambda n:1 if n<=100 else 2 if n<=1000 else 3)
df['Third_Rank_Gap']=df['Rank_Third_Week'].apply(lambda n:1 if n<=100 else 2 if n<=1000 else 3)
df['Forth_Rank_Gap']=df['Rank_Forth_Week'].apply(lambda n:1 if n<=100 else 2 if n<=1000 else 3)
df['Final_Rank_Gap']=df['Rank_Final_Week'].apply(lambda n:1 if n<=100 else 2 if n<=1000 else 3)
df['First']=df['First_Week']-df['Second_Week']
df['Second']=df['Second_Week']-df['Third_Week']
df['Third']=df['Third_Week']-df['Forth_Week']
df['First_tendency']=(df.Rank_Second_Week <df.Rank_First_Week).apply(lambda n : 'increase'if n ==True else 'decrease')
df['Second_tendency']=(df.Rank_Third_Week <df.Rank_Second_Week).apply(lambda n : 'increase'if n ==True else 'decrease')
df['Third_tendency']=(df.Rank_Forth_Week <df.Rank_Third_Week).apply(lambda n : 'increase'if n ==True else 'decrease')
df['Forth_tendency']=(df.Final_Week <df.Rank_Forth_Week).apply(lambda n : 'increase'if n ==True else 'decrease')
df.head(10)
```

```python
from __future__ import print_function

import os
import subprocess

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
def visualize_tree(tree, feature_names):
    """Create tree png using graphviz.

    Args
    ----
    tree -- scikit-learn DecsisionTree.
    feature_names -- list of feature names.
    """
    with open("dt.dot", 'w') as f:
        export_graphviz(tree, out_file=f,
                        feature_names=feature_names)

    command = ["dot", "-Tpng", "dt.dot", "-o", "dt.png"]
    try:
        subprocess.check_call(command)
    except:
        exit("Could not run dot, ie graphviz, to "
             "produce visualization")
```
