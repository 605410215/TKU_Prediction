```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
df2=pd.DataFrame(columns=['msno','song_id','song_length','played_length','day','artist_id','hear_percent','hear_gap'])
df2=df2.groupby(['artist_id','day']).count()
size=0
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        #df['week']=df.day.apply(lambda n : 'First_Week' if n<=4 else 'Second_Week' if n<=11 else 'Third_Week' if n<=18 else 'Forth_Week' if n <= 25 else 'Final_Week')
        df=df.groupby(['artist_id','day']).count()
        df2=df2.add(df,fill_value=0)       
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
print ('Data_size:%d'%(size))
```

```python
df2=df2.reset_index()
df2.head()
```

```python
#df2=df2.reset_index()
df2=df2.drop(['hear_gap','song_length','played_length','song_id','hear_percent','hear_gap2'],axis=1)
df2['artist_id']=df2['artist_id'].astype(str)
df2['artist_id']=df2['artist_id'].apply(lambda n:n[:n.index('.')])
df2.rename(columns={'msno': 'number'}, inplace=True)
df2['number']=df2['number'].astype(str)
df2['number']=df2['number'].apply(lambda n:n[:n.index('.')])
df2['number']=df2['number'].astype('int64')
'''
df2['Rank_number'] = df2['number'].rank(ascending=False,method='dense')
df2['Rank_number']=df2['Rank_number'].astype(str)
df2['Rank_number']=df2['Rank_number'].apply(lambda n:n[:n.index('.')])
df2['Rank_number']=df2['Rank_number'].astype('int64')
'''
df2.head(10)
```

```python
df3=df2.pivot(index='artist_id', columns='day', values='number').reset_index().rename_axis(None, axis=1)
df3.head(10)
```

```python
df4=df3.fillna(0)
for h in range(1,30):
    df4[str(h)+'_'+str(h+1)]=(df4[h+1]-df4[h])/((h+1)-h)
    df4[str(h)+'_'+str(h+1)]=df4[str(h)+'_'+str(h+1)].astype(str)
    df4[str(h)+'_'+str(h+1)]=df4[str(h)+'_'+str(h+1)].apply(lambda n:n[:n.index('.')])
    df4[str(h)+'_'+str(h+1)]=df4[str(h)+'_'+str(h+1)].astype('int64')
    print(h,end=',')
```

```python
for e in range(1,31):
    df4=df4.drop(e,axis=1)
```
