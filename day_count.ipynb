```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

```python
df2=pd.DataFrame(columns=['msno','song_id','song_length','played_length','day','artist_id','hear_percent','hear_gap'])
df2=df2.groupby(['artist_id','day'])['msno'].count()
size=0
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='a'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df['day']=df.day.apply(lambda n : 'First_Week_People' if n<=4 else 'Second_Week_People' if n<=11 else 'Third_Week_People' if n<=18 else 'Forth_Week_People' if n <= 25 else 'Final_Week_People')
        #df['week']=df.day.apply(lambda n : 'First_Week' if n<=4 else 'Second_Week' if n<=11 else 'Third_Week' if n<=18 else 'Forth_Week' if n <= 25 else 'Final_Week')
        df=df.groupby(['artist_id','day'])['msno'].count()
        df2=df2.add(df,fill_value=0)       
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
print ('Data_size:%d'%(size))
```

```python
df2.reset_index()
```

```python
df3=df2.reset_index()
df3.rename(columns={'msno': 'number'}, inplace=True)
df3['artist_id']=df3['artist_id'].astype(str)
df3['artist_id']=df3['artist_id'].apply(lambda n:n[:n.index('.')])
df3['number']=df3['number'].astype(str)
df3['number']=df3['number'].apply(lambda n:n[:n.index('.')])
df3['number']=df3['number'].astype('int64')
df3
```

```python
df4=df3.pivot(index='artist_id', columns='day', values='number').reset_index().rename_axis(None, axis=1)
df4=df4.fillna(0)
df4.head(10)
```

```python
df4['Ttotal_week_people']=df4['First_Week_People']+df4['Second_Week_People']+df4['Third_Week_People']+df4['Forth_Week_People']
df4['Avg_week_people']=df4['Ttotal_week_people']/4
df4
```
